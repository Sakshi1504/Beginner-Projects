```python
#squeeze parameters - converts the column into series object and it doesn't stay in form of dataframe
df=pd.read_csv("Top 10 State Visit.csv", usecols=['year'], squeeze=True)
df
```

```python
#skiprows/nrows Parameter
df=pd.read_csv("Top 10 State Visit.csv", skiprows=[0,3]) #function can also be used inside skiprows in form of lambda
df
```

```python
#nrows will result number of rows mentioned
df=pd.read_csv("Top 10 State Visit.csv", nrows=3)
df
```

```python
#encoding - some kaggle datasets don't have encoding UTF-8
df=pd.read_csv("Top 10 State Visit.csv", encoding='UTF-8') #jis file ki jo encoding hai wo mention krte hai to load it
df
```

```python
#skip bad lines - skips lines with erroneaous data / extra data in rows not aligned with columns parser error
df=pd.read_csv("error_row.csv", error_bad_lines=True)
df
```

```python
#dtype - changing data types of columns
df=pd.read_csv("Top 10 State Visit.csv", dtype={'year':int}).info()
df
```

```python
#Handling of dates - by default dates are imported as String
df=pd.read_csv("Top 10 State Visit.csv",parse_dates=['year'])
#it will start treating string column of dates in date format on which dates operations can be applied
df
```

```python
def spc(name):
  if name == 'Tamil Nadu':
    return 'TN'
  else:
    return 'NA'
```

```python
#Convertors - transform the data while loading it itself - applies function on specific column
df=pd.read_csv("Top 10 State Visit.csv",converters={'top1_state':spc})
df
```

```python
#na values parameter - where null values are not easily identifiable - like dash in place of Nan/null
#so we can specify what kind of values do we need to consider as Nan
df=pd.read_csv("Top 10 State Visit.csv", na_values= 'Tamil Nadu')
df
```

```python
#Loading a huge dataset in chunks - data can be loaded into chunks and can then be populated as smaller chunks
dfs=pd.read_csv("Top 10 State Visit.csv", chunksize=3)
df
```
